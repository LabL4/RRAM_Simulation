```python
from RRAM import *
from RRAM import Montecarlo
import numpy as np
```

```python
# Creo el vector con todos los factores que quiero probar que es desde 0.0001 hasta 10000 en salto de un orden de magnitud
factors = np.logspace(1, 5, num=5)
print(factors)
```

```python
# Hago un ciclo para cada factor realizar el montecarlo y guardar los datos con el factor correspondiente

for factor in factors:

    # Excribo el factor que se esta calculando
    print('Factor: {}'.format(factor))
    data = Montecarlo.funcion_montecarlo(10000, 3, 3, factor)

    # Suponiendo que 'data' es un array de NumPy que ya contiene tus datos
    data_filtrados = np.array([fila for fila in data if fila[-1] != 0.0])

    np.savetxt('Recombinacion_data_{}.csv'.format(factor), data_filtrados,  fmt=['%d', '%.5f', '%d', '%d', '%f', '%.6e', '%.1f'],
               header='paso, tiempo simulacion, x, y, probabilidad recombinacion, velocidad, funcion a trozos', comments='', delimiter=', ')
```
