```python
import matplotlib.pyplot as plt
import scipy.constants as cte
import numpy as np
import math

from RRAM import Constants as cte
```

```python
t_0 = cte.t_0
beta_0 = cte.beta_0
E_a = cte.E_a
cte_red = cte.cte_red
L_p = cte.L_p
cte_red = cte.cte_red

paso_temporal = 1e-3
k_b_ev = cte.k_b_ev
temp = 300
velocidad = 5.525555900642030425e-07

pos_x = np.linspace(1, 40, 40)
prob_recom = np.zeros(40)
```

```python

# Calculo la probabilidad de recombinación en equilibrio

prob_in_equilibrio = (paso_temporal * t_0) * (math.exp(-E_a / (k_b_ev * temp)))

for i in range(40):
    funcion_trozos = cte.DifussiveBehaviour(i, velocidad, paso_temporal, cte_red)
    exp_beta = math.exp(-(paso_temporal * velocidad) / L_p) * beta_0
    prob_recom[i] = prob_in_equilibrio * exp_beta

prob_recom
```
