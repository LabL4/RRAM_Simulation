```python
import pandas as pd
import numpy as np
import time
```

```python

# Crear un DataFrame de muestra
data = pd.DataFrame(np.random.rand(100000, 10), columns=[f'col{i}' for i in range(10)])

# Guardar en un archivo de texto
start_time = time.time()
data.to_csv('data.txt', sep='\t', index=False)
print(f"Guardar en txt: {time.time() - start_time} segundos")

# Guardar en un archivo de Excel
start_time = time.time()
data.to_excel('data.xlsx', index=False)
print(f"Guardar en excel: {time.time() - start_time} segundos")
```
