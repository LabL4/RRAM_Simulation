```python
import matplotlib.pyplot as plt

def config_ax(ax):
    ax.grid(which='major', color='#DDDDDD', linewidth=0.8, zorder=-1)
    ax.grid(which='minor', color='#DEDEDE', linestyle=':', linewidth=0.5, zorder=-1)
    ax.minorticks_on()
    ax.tick_params(axis='both', which='both', direction='in', top=True, right=True)


def setup_plt(plt, latex=True, scaling=1):
    plt.rcParams.update(
        {
            "pgf.texsystem": "pdflatex",
            "text.usetex": latex,
            "font.family": "fourier",
            "text.latex.preamble": "\n".join([
                r"\usepackage[utf8]{inputenc}",
                r"\usepackage[T1]{fontenc}",
                r"\usepackage{siunitx}",
            ])
        }
    )

    SMALL_SIZE = 8 * scaling
    MEDIUM_SIZE = 10 * scaling
    BIGGER_SIZE = 11 * scaling

    plt.rc('font', size=SMALL_SIZE)
    plt.rc('axes', titlesize=SMALL_SIZE)
    plt.rc('axes', labelsize=MEDIUM_SIZE)
    plt.rc('xtick', labelsize=SMALL_SIZE)
    plt.rc('ytick', labelsize=SMALL_SIZE)
    plt.rc('legend', fontsize=SMALL_SIZE)
    plt.rc('figure', titlesize=BIGGER_SIZE)
    plt.rc('axes', titlesize=BIGGER_SIZE*1.05)


# === Configuración de estilo LaTeX y plot ===
setup_plt(plt, latex=True, scaling=2)

# === Plot SET ===
fig_set, ax_set = plt.subplots(figsize=(7, 6))
config_ax(ax_set)
```

```python
from pathlib import Path

from matplotlib import lines
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import math

# === CONFIGURACIÓN GENERAL ===
base_path = Path("Results")
exp_path_n = Path("Datos_Experimentales/Ciclos_Experimentales/Cycle_n_1000.txt")
exp_path_p = Path("Datos_Experimentales/Ciclos_Experimentales/Cycle_p_1000.txt")
total_simulations = 2

E_a_base = 1.05
E_a_values  = [0.95*E_a_base,1.05*E_a_base]

E_r_base = 1.05
E_r_values = [0.95*E_a_base,1.05*E_a_base]

figures_per_panel = 2

# Carpeta de guardado
figures_path = base_path / "Figures"
figures_path.mkdir(parents=True, exist_ok=True)
```

```python
total_panels = math.ceil(total_simulations / figures_per_panel)

sampling_rate_sim = 250  # Muestreo general

# RANGOS PARA PPSSET
dense_ranges = [
    (0.3, 0.45, 200),
    (0.45, 0.9, 2),
    (0.9, 1.1, 200)
]

# === Medidas experimentales ===
try:
    data_exp_n = np.loadtxt(exp_path_n)
    x_exp_n = data_exp_n[:, 0]
    y_exp_n = np.abs(data_exp_n[:, 1])
    print(f"✅ Medidas experimentales (n) cargadas completas.")
except Exception as e:
    print(f"❌ Error leyendo {exp_path_n}: {e}")
    x_exp_n, y_exp_n = None, None

try:
    data_exp_p = np.loadtxt(exp_path_p)
    x_exp_p = data_exp_p[:, 0]
    y_exp_p = np.abs(data_exp_p[:, 1])
    print(f"✅ Medidas experimentales (p) cargadas completas.")
except Exception as e:
    print(f"❌ Error leyendo {exp_path_p}: {e}")
    x_exp_p, y_exp_p = None, None

# === Marcadores para cada simulación ===
markers = ['o', 's', '^', 'v', 'D', 'P', '*', 'X', 'h', '<', '>', '|', '_', 'H']

for panel_idx in range(total_panels):
    Ea_value = E_a_values[panel_idx % len(E_a_values)]
    fig, ax = plt.subplots(figsize=(10, 8))

    for subplot_idx in range(figures_per_panel):
        sim_num = panel_idx * figures_per_panel + subplot_idx + 1
        if sim_num > total_simulations:
            continue

        Er_value = E_r_values[subplot_idx % len(E_r_values)]
        sim_path = base_path / f"simulation_{sim_num}"

        grouped_patterns = [
            ("set/resultados_pp_set_{}.csv", "Set", True),
            ("set/Resultados_sp_set_{}.csv", "Set", False),
            ("reset/resultados_pp_reset_{}.csv", "Reset", False),
            ("reset/resultados_sp_reset_{}.csv", "Reset", False)
        ]

        x_all, y_all = [], []
        data_found = False

        for pattern, label, is_pp_set in grouped_patterns:
            file_path = sim_path / pattern.format(sim_num)
            if file_path.exists():
                try:
                    df = pd.read_csv(file_path)
                    if df.shape[1] >= 3:
                        x_data = df.iloc[:, 1]
                        y_data = np.abs(df.iloc[:, 2])

                        if is_pp_set:
                            parts_x, parts_y = [], []
                            used_indices = pd.Series(False, index=x_data.index)

                            for x_min, x_max, rate in dense_ranges:
                                mask = (x_data >= x_min) & (x_data <= x_max)
                                idx = mask[mask].index[::rate]
                                parts_x.append(x_data.loc[idx])
                                parts_y.append(y_data.loc[idx])
                                used_indices.loc[idx] = True

                            outside_idx = (~used_indices)[~used_indices].index[::sampling_rate_sim]
                            parts_x.append(x_data.loc[outside_idx])
                            parts_y.append(y_data.loc[outside_idx])

                            x_combined = pd.concat(parts_x, ignore_index=True)
                            y_combined = pd.concat(parts_y, ignore_index=True)

                        else:
                            x_combined = x_data.iloc[::sampling_rate_sim]
                            y_combined = y_data.iloc[::sampling_rate_sim]

                        x_all.append(x_combined)
                        y_all.append(y_combined)
                        data_found = True
                    else:
                        print(f"⚠️ {file_path} tiene menos de 3 columnas.")
                except Exception as e:
                    print(f"❌ Error leyendo {file_path}: {e}")
            else:
                print(f"⚠️ Archivo no encontrado: {file_path}")

        if data_found:
            x_concat = pd.concat(x_all, ignore_index=True)
            y_concat = pd.concat(y_all, ignore_index=True)
            legend_label = r"$E_r = " + f"{Er_value:.3f}" + r"$"
            marker_style = markers[(sim_num - 1) % len(markers)]
            ax.scatter(x_concat, y_concat, s=10, alpha=1, label=legend_label, marker=marker_style)
        else:
            print(f"ℹ️ Simulación {sim_num} sin datos, no se representará.")

    # === Medidas experimentales completas ===
    if x_exp_n is not None and y_exp_n is not None:
        ax.plot(x_exp_n, y_exp_n, color='black', linewidth=2.5, alpha=1, label="Medidas Experimentales", linestyle='--')
    if x_exp_p is not None and y_exp_p is not None:
        ax.plot(x_exp_p, y_exp_p, color='black', linewidth=2.5, alpha=1, linestyle='--')

    ax.set_title(
        rf"$E_a = {Er_value:.3f}$ | Panel {panel_idx + 1}: Sim {panel_idx * figures_per_panel + 1} - {min((panel_idx + 1) * figures_per_panel, total_simulations)}"
    )
    ax.set_xlabel('Voltage [V]')
    ax.set_ylabel('Current [A]')
    ax.set_yscale("log")
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)

    ax.legend(fontsize=12, loc='best', ncol=2)

    plt.tight_layout()
    save_name = figures_path / f"panel_COMBINADO_panel_{panel_idx+1}_Ea_{Er_value:.3f}.png"
    plt.savefig(save_name, dpi=300)
    print(f"✅ Panel combinado guardado en {save_name}")
    plt.close(fig)
```
