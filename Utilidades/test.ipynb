```python
# Represento el estado de los oxígenos cada 50 pasos
num_representar = 50
num_simulation = 0
paso_guardar = 50
num_pasos = 5750
```

```python
with open(ruta_raiz + f'Results/reset/Oxygen_pp_reset_{num_simulation}.pkl', 'rb') as f:
    oxygen_pp_reset = pickle.load(f)
with open(ruta_raiz + f'Results/reset/Oxygen_sp_reset_{num_simulation}.pkl', 'rb') as f:
    oxygen_sp_reset = pickle.load(f)

vector_guardar = np.arange(0, num_pasos, paso_guardar)

print("Representando el estado de los oxígenos")
for i in vector_guardar:
    k = i // paso_guardar
    pplt.RepresentateState(oxygen_pp_reset[i], ruta_raiz + f'Results/Figures/Oxygen_pp_reset_{num_simulation}_{k}.png')
    pplt.RepresentateState(oxygen_sp_reset[i], ruta_raiz + f'Results/Figures/Oxygen_sp_reset_{num_simulation}_{k}.png')
print("Acabado")
```

```python
with open(ruta_raiz + f'Results/reset/Configurations_pp_reset_{num_simulation}.pkl', 'rb') as f:
    Configurations_pp_reset = pickle.load(f)
with open(ruta_raiz + f'Results/reset/Configurations_sp_reset_{num_simulation}.pkl', 'rb') as f:
    Configurations_sp_reset = pickle.load(f)

vector_guardar = np.arange(0, num_pasos, paso_guardar)

print("Representando el estado de las vacantes")
for i in vector_guardar:
    k = i // paso_guardar
    pplt.RepresentateState(Configurations_pp_reset[i], ruta_raiz + f'Results/Figures/Configurations_pp_reset_{num_simulation}_{k}.png')
    pplt.RepresentateState(Configurations_sp_reset[i], ruta_raiz + f'Results/Figures/Configurations_sp_reset_{num_simulation}_{k}.png')
```
