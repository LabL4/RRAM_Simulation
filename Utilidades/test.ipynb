```python
import numpy as np
import matplotlib.pyplot as plt
import sys
```

```python
ruta_raiz = 'C:/Users/Usuario/Documents/GitHub/RRAM_Simulation/'
# ruta_raiz = '/Users/antonio_lopez_torres/Documents/GitHub/RRAM_Simulation/' # Ruta en el mac
sys.path.append(ruta_raiz)
```

```python
# Cargar los datos desde el archivo de texto, omitiendo la primera fila (encabezado)
data = np.loadtxt(ruta_raiz + 'Results/set/Vacantes_resistencia_0.txt', skiprows=1)
# Asumimos que el archivo tiene tres columnas: tiempo de simulación, voltaje y número de vacantes
simulation_time = data[:, 0]
voltaje = data[:, 1]
resistencia = data[:, 3]
num_vacantes = data[:, 2]
```

```python
# Ejemplo de representación gráfica de voltaje y número de vacantes en función del tiempo de simulación
plt.figure(figsize=(20, 5))

ticks_x = np.arange(0, 2.51, 0.1)  # Desde 0 hasta 10, con un intervalo de 0.5


# Graficar voltaje vs tiempo de simulación
plt.subplot(1, 2, 1)
plt.plot(voltaje, resistencia, label="Resistencia", color="blue")
plt.xlabel("Voltaje")
plt.ylabel("Resistencia")
plt.title("Resistencia vs Voltaje")
plt.legend()
plt.xticks(ticks_x)

# Graficar número de vacantes vs Voltaje")
plt.subplot(1, 2, 2)
plt.plot(voltaje, num_vacantes, label="Número de Vacantes", color="green")
plt.xlabel("Voltaje")
plt.ylabel("Número de Vacantes")
plt.title("Vacantes vs Voltaje")
plt.legend()
plt.xticks(ticks_x)

# Mostrar los gráficos
plt.tight_layout()
plt.show()
```
