```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Leer datos del fichero txt
data = np.loadtxt('datos.txt')

# Asumimos que los datos están en dos columnas: x e y
x = data[:, 0]
y = data[:, 1]

# Crear la gráfica scatter
plt.scatter(x, y)

# Configurar el eje y en escala logarítmica
plt.yscale('log')

# Añadir etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y (Escala Logarítmica)')
plt.title('Gráfica Scatter con Eje Y en Escala Logarítmica')

# Mostrar la gráfica
plt.show()
```

```python
import os
import matplotlib.pyplot as plt
import numpy as np

# Ruta proporcionada
ruta = 'C:/Users/Usuario/Desktop/Datos_ciclos_RRAM/Ciclos_Set'

# Obtener lista de archivos en la ruta que no contienen 'n' en su nombre
archivos = [f for f in os.listdir(ruta) if os.path.isfile(os.path.join(ruta, f)) and 'n' not in f]

# Inicializar listas para almacenar datos
x_total = []
y_total = []

# Leer y procesar cada archivo
for archivo in archivos:
    data = np.loadtxt(os.path.join(ruta, archivo))
    x_total.extend(data[:, 0])
    y_total.extend(data[:, 1])

# Crear la gráfica scatter
plt.scatter(x_total, y_total)

# Configurar el eje y en escala logarítmica
plt.yscale('log')

# Añadir etiquetas y título
plt.xlabel('Voltaje [V]')
plt.ylabel('Intensidad [A]')
plt.title('Corriente de Set en función del Voltaje')

# Mostrar la gráfica
plt.show()
```
