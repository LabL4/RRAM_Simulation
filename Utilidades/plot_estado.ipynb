```python
from RRAM import utils, Representate
from pathlib import Path


import numpy as np
import pickle
```

```python
def cagar_estado(state_path: Path):
    # Asegúrate de abrir el archivo correctamente usando el parámetro state_path
    with open(state_path, "rb") as file:
        state = pickle.load(file)
    return state
```

```python
num_simulation = 15
fig_voltage = -1.1256
ruta_pkl = (
    Path.cwd()
    / "Results"
    / f"simulation_{num_simulation}"
    / "reset"
    / f"filamento_1_roto_reset_{num_simulation}.pkl"
)

state = cagar_estado(ruta_pkl)

ruta_fig = Path.cwd() / f"Filamento_1_roto_reset_{num_simulation}.png"
```

```python
Representate.RepresentateState(state, fig_voltage, str(ruta_fig))
```
