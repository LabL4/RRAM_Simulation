```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os

# Ahora importa el módulo
from RRAM.Representate import config_ax, setup_paper_plt
    
setup_paper_plt(plt, latex=True, scaling=2)
```

##### Cargo los datos experimentales

```python
# Rutas de los archivos experimentales
ruta_archivo_set = os.getcwd() + "/Datos_Experimentales/Medidas_Experimentales_RRAM/Cycle_p_1000.txt"
ruta_archivo_reset = os.getcwd() + "/Datos_Experimentales/Medidas_Experimentales_RRAM/Cycle_n_1000.txt"

df_exp_set = np.genfromtxt(
    ruta_archivo_set,
    dtype=[
        ("voltage", "f8"),  # String Unicode de máx 50 caracteres
        ("current", "f8"),  # Float de 64 bits
        ("time", "f8"),  # Float de 64 bits
    ],
    encoding="utf-8",
)

df_exp_reset = np.genfromtxt(
    ruta_archivo_reset,
    dtype=[
        ("voltage", "f8"),  # String Unicode de máx 50 caracteres
        ("current", "f8"),  # Float de 64 bits
        ("time", "f8"),  # Float de 64 bits
    ],
    encoding="utf-8",
)

# Crear un DataFrame para una mejor visualización
df_exp_set = pd.DataFrame(df_exp_set)
df_exp_reset = pd.DataFrame(df_exp_reset)
```

##### Cargo los resultados del modelo de Stanford que mejor ajustan

```python
# Rutas de los archivos experimentales
ruta_archivo_set = os.getcwd() + "/Stanford_Data/simulated_data_set.txt"
ruta_archivo_reset = os.getcwd() + "/Stanford_Data/simulated_data_reset.txt"

df_sim_set = np.genfromtxt(
    ruta_archivo_set,
    dtype=[
        ("time", "f8"),  # String Unicode de máx 50 caracteres
        ("voltage", "f8"),  # Float de 64 bits
        ("current", "f8"),  # Float de 64 bits
    ],
    encoding="utf-8",
)

df_sim_reset = np.genfromtxt(
    ruta_archivo_reset,
    dtype=[
        ("time", "f8"),  # String Unicode de máx 50 caracteres
        ("voltage", "f8"),  # Float de 64 bits
        ("current", "f8"),  # Float de 64 bits
    ],
    encoding="utf-8",
)

# Crear un DataFrame para una mejor visualización
df_sim_set = pd.DataFrame(df_sim_set)
df_sim_reset = pd.DataFrame(df_sim_reset)
```

##### Represento los datos y la curva ajustada del modelo de Stanford

```python
setup_paper_plt(plt, latex=True, scaling=2)

fig, axes = plt.subplots(1, 1, figsize=(12, 9))
config_ax(axes)

voltage_exp = np.concatenate((np.array(df_exp_set["voltage"]), np.array(df_exp_reset["voltage"])))
current_exp = np.concatenate((np.array(df_exp_set["current"]), np.array(df_exp_reset["current"])))

voltage_sim = np.concatenate((np.array(df_sim_set["voltage"]), np.array(df_sim_reset["voltage"])))
current_sim = np.concatenate((np.array(df_sim_set["current"]), np.array(df_sim_reset["current"])))

axes.plot(voltage_exp, abs(current_exp), color="black", label="Experimental data", linewidth=2.5, linestyle="dashed")
axes.plot(
    voltage_sim,
    abs(current_sim),
    color="blue",
    label="Stanford model data",
    linewidth=2,
)

# Leyenda ajustada en la parte inferior izquierda
axes.legend(
        labelspacing=0.3,
        handletextpad=0.2,
        handlelength=1.0,
        borderaxespad=0.2,
        loc="best",
    )
    
axes.set_yscale("log")
axes.set_xlabel(r"Voltage (\si{\volt})", fontsize=32)
axes.set_ylabel(r"Current (\si{\ampere)", fontsize=32)

fig.savefig(
    os.getcwd() + "/Ajuste_modelo_stanford.pdf",
    bbox_inches="tight",
    dpi=300,
)
plt.close(fig)  # Cierra para liberar memoria
```

##### Modelo de Stanford con distintas $E_a$ para el SET

```python
setup_paper_plt(plt, latex=True, scaling=2)

fig, axes = plt.subplots(1, 1, figsize=(12, 9))
config_ax(axes)

voltage_exp = np.array(df_exp_set["voltage"])
current_exp = np.array(df_exp_set["current"])

axes.plot(
    voltage_exp,
    abs(current_exp),
    color="black",
    label="Experimental data",
    linewidth=3,
    linestyle="dashed",
)


for factor in [0.99, 0.95, 0.90, 0.85]:  # cuidado que el ultimo valor NO lo coge

    E_a = round(factor * 0.6,3)  # eV
    # Rutas de los archivos experimentales
    ruta_archivo_set = (
        os.getcwd()
        + f"/Stanford_Data/simulated_data_set_{factor}.txt"
    )
    
    df_set = np.genfromtxt(
        ruta_archivo_set,
        dtype=[
            ("time", "f8"),  # Float de 64 bits
            ("voltage", "f8"),  # String Unicode de máx 50 caracteres
            ("current", "f8"),  # Float de 64 bits
        ],
        encoding="utf-8",
    )
        
    voltage_exp = np.concatenate((np.array(df_set["voltage"]), np.array(df_exp_reset["voltage"])))
    current_exp = np.concatenate((np.array(df_set["current"]), np.array(df_exp_reset["current"])))
    
    
    axes.plot(
        df_set["voltage"],
        abs(df_set["current"]),
        label = r"$E_a = " + str(factor) + r" E_a\,\, (\si{e\volt}) $",
        linewidth=2.5,
    )

    axes.set_yscale("log")
    axes.set_xlabel(r"Voltage (\si{\volt})", fontsize=32)
    axes.set_ylabel(r"Current (\si{\ampere)", fontsize=32)
    

# Leyenda ajustada en la parte inferior izquierda
axes.legend(
    labelspacing=0.3,
    handletextpad=0.2,
    handlelength=1.0,
    borderaxespad=0.2,
    loc="lower right",
)

fig.savefig(
    os.getcwd() + "/Set_modelo_stanford.pdf",
    bbox_inches="tight",
    dpi=300,
)
plt.close(fig)  # Cierra para liberar memoria
```

##### Simulación para distintos $E_a$ en el set

```python
carpeta_results = os.getcwd() + "/Results/"

setup_paper_plt(plt, latex=True, scaling=2)

fig, axes = plt.subplots(1, 1, figsize=(12, 9))
config_ax(axes)

voltage_exp = np.array(df_exp_set["voltage"])
current_exp = np.array(df_exp_set["current"])

axes.plot(
    voltage_exp,
    abs(current_exp),
    color="black",
    label="Experimental data",
    linewidth=3,
    linestyle="dashed",
)

axes.set_yscale("log")
axes.set_xlabel(r"Voltage (\si{\volt})", fontsize=32)
axes.set_ylabel(r"Current (\si{\ampere)", fontsize=32)

E_a = 1.01  # eV

for num_simulation, factor in enumerate([0.99, 0.95, 0.90, 0.85]):  # cuidado que el ultimo valor NO lo coge
    print(f"Processing simulation {num_simulation + 1} with factor {factor}")
    E_a = round(factor * E_a, 3)  # eV
    set_simulation_path = os.path.join(carpeta_results, f"simulation_{num_simulation + 1}/set/")
    
    try:
        # Cargo los datos de las simulaciones de SET
        data_path_pp_set = set_simulation_path + f"resultados_pp_set_{num_simulation + 1}.csv"
        data_path_sp_set = set_simulation_path + f"resultados_sp_set_{num_simulation + 1}.csv"
        
        df_pset = pd.read_csv(data_path_pp_set, dtype=float)
        df_sset = pd.read_csv(data_path_sp_set, dtype=float)
    except FileNotFoundError:
        print(f"Files not found for simulation {num_simulation + 1}. Skipping...")
        continue

    i_ps = np.array(df_pset["Intensidad [A]"])
    i_ss = np.array(df_sset["Intensidad [A]"])

    v_ps = np.array(df_pset["Voltaje [V]"])
    v_ss = np.array(df_sset["Voltaje [V]"])

    # Uno en un solo array las intensidades y los voltajes de las curvas de SET
    i_set = np.concatenate((i_ps, i_ss))
    v_set = np.concatenate((v_ps, v_ss))

    axes.scatter(v_set, abs(i_set), label=r"$E_a = " + str(factor) + r" E_a\,\, (\si{e\volt}) $", linewidth=1, s=15)


# Leyenda ajustada en la parte inferior izquierda
axes.legend(
    labelspacing=0.3,
    handletextpad=0.2,
    handlelength=1.0,
    borderaxespad=0.2,
    loc="lower right",
)

fig.savefig(
    os.getcwd() + "/Set_simulacion.pdf",
    bbox_inches="tight",
    dpi=300,
)
plt.close(fig)  # Cierra para liberar memoria
```
