```python
import matplotlib.pyplot as plt

def config_ax(ax):
    ax.grid(which='major', color='#DDDDDD', linewidth=0.8, zorder=-1)
    ax.grid(which='minor', color='#DEDEDE', linestyle=':', linewidth=0.5, zorder=-1)
    ax.minorticks_on()
    ax.tick_params(axis='both', which='both', direction='in', top=True, right=True)


def setup_plt(plt, latex=True, scaling=1):
    plt.rcParams.update(
        {
            "pgf.texsystem": "pdflatex",
            "text.usetex": latex,
            "font.family": "fourier",
            "text.latex.preamble": "\n".join([
                r"\usepackage[utf8]{inputenc}",
                r"\usepackage[T1]{fontenc}",
                r"\usepackage{siunitx}",
            ])
        }
    )

    SMALL_SIZE = 8 * scaling
    MEDIUM_SIZE = 10 * scaling
    BIGGER_SIZE = 11 * scaling

    plt.rc('font', size=SMALL_SIZE)
    plt.rc('axes', titlesize=SMALL_SIZE)
    plt.rc('axes', labelsize=MEDIUM_SIZE)
    plt.rc('xtick', labelsize=SMALL_SIZE)
    plt.rc('ytick', labelsize=SMALL_SIZE)
    plt.rc('legend', fontsize=SMALL_SIZE)
    plt.rc('figure', titlesize=BIGGER_SIZE)
    plt.rc('axes', titlesize=BIGGER_SIZE*1.05)


# === Configuración de estilo LaTeX y plot ===
setup_plt(plt, latex=True, scaling=2)
```

```python
# Diccionarios separados
data_reset = {}
data_reset = {}

# === Cargar archivos txt de set/reset ===
for num_sim in range(1, total_simulations + 1):
    sim_path = base_path / f"simulation_{num_sim}"
    
    # RESET
    reset_path = sim_path / "reset"
    archivos_reset = [
        (reset_path / f"g_pp_reset_{num_sim}.txt", f"reset_pp_{num_sim}"),
        (reset_path / f"g_sp_reset_{num_sim}.txt", f"reset_sp_{num_sim}")
    ]
    
    for path_archivo, nombre in archivos_reset:
        if path_archivo.exists():
            data = np.loadtxt(path_archivo, delimiter=',')
            data_reset[nombre] = data
            print(f"✅ RESET: Cargado {nombre} con forma {data.shape}")
        else:
            print(f"⚠️ RESET: Archivo no encontrado: {path_archivo}")

    # SET
    set_path = sim_path / "set"
    archivos_set = [
        (set_path / f"g_pp_set_{num_sim}.txt", f"set_pp_{num_sim}"),
        (set_path / f"g_sp_set_{num_sim}.txt", f"set_sp_{num_sim}")
    ]
    
    for path_archivo, nombre in archivos_set:
        if path_archivo.exists():
            data = np.loadtxt(path_archivo, delimiter=',')
            data_reset[nombre] = data
            print(f"✅ SET: Cargado {nombre} con forma {data.shape}")
        else:
            print(f"⚠️ SET: Archivo no encontrado: {path_archivo}")
```

```python
base_path = "c:/Users/jimdo/Documents/GitHub/RRAM_Simulation"


num_simulacion = 1

ruta_raiz = os.path.normpath(base_path + f'/Results/simulation_{num_simulacion}/')
g_figure_path = os.path.normpath(base_path + '/Results/Figures_g_evolution/')

# === Cargar archivos txt de set ===
g_data_path_set = os.path.normpath(ruta_raiz + f'/set/g_set_{num_simulacion}.txt')
results_path_set = os.path.normpath(ruta_raiz + f'/set/Resultados_set_{num_simulacion}.csv')

data_g_set = io.leer_txt_as_csv(g_data_path_set, header=None) # type: ignore
data_set = io.leer_csv(results_path_set)

# Filtro para obtener las g donde las columnas NO son simepre 40
g_filtered_set =  data_g_set.loc[:, data_g_set.nunique(dropna=False) > 2]
```

```python
# === Crear figura con tamaño para póster ===
fig_set, ax_set = plt.subplots(figsize=(12, 9))
config_ax(ax_set)

yticks_custom = np.arange(40, -1, -5)  # De 40 a 0 (incluido), en pasos de -5
yticks_custom = yticks_custom * (10/40)
ax_set.yaxis.set_major_locator(FixedLocator(yticks_custom))

print("yticks_custom:", yticks_custom)

# === Graficar cada columna del DataFrame ===
for columna in g_filtered_set.columns:
    ax_set.plot(data_set['# Tiempo simulacion [s]'], g_filtered_set[columna]* (10/40), label=f" Fila {columna}")
    
# === Etiquetas y título ===
ax_set.set_xlabel(r"Tiempo (\SI{}{\second})")
ax_set.set_ylabel(r"$g$ (\SI{}{\nano\meter})")
ax_set.set_title("Evolución de $g$ durante el set", pad=15)

# === Leyenda y estilo ===
ax_set.legend(loc='best', fontsize='small')

# === Mostrar o guardar la figura ===
plt.close(fig_set)
fig_set.savefig(g_figure_path + "/g_completa_set.png", bbox_inches='tight', dpi=300)

# creo el vector de g medio

g_medio = g_filtered_set.mean(axis=1).to_numpy() * (10/40)  # Ajuste de escala

# Represento el g medio 
fig_g_medio, ax_g_medio_set= plt.subplots(figsize=(12, 9))
config_ax(ax_g_medio_set)

ax_g_medio_set.plot(data_set["# Tiempo simulacion [s]"], g_medio, label="Valor medio de g")

# === Etiquetas y título ===
ax_g_medio_set.set_xlabel(r"Tiempo (\SI{}{\second})")
ax_g_medio_set.set_ylabel(r"$g$ (\SI{}{\nano\meter})")
ax_g_medio_set.set_title("Evolución de $g$ medio durante el set", pad=15)

# === Leyenda y estilo ===
ax_g_medio_set.legend(loc='best', fontsize='small')

# === Mostrar o guardar la figura ===
# plt.show()
fig_g_medio.savefig(g_figure_path + "/g_medio_set.png", bbox_inches='tight', dpi=300)

plt.close(fig_g_medio)
```
