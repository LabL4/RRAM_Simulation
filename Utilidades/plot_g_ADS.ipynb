```python
import numpy as np
import matplotlib.pyplot as plt

# Load data from files teniendo en cuenta NO leer la priemra fila
g_reset_data = np.loadtxt(
    r"C:/Users/jimdo/Documents/GitHub/RRAM_Simulation/ADS/g_reset_vs_V.txt", skiprows=1
)
g_set_data = np.loadtxt(
    r"C:/Users/jimdo/Documents/GitHub/RRAM_Simulation/ADS/g_set_vs_V.txt", skiprows=1
)

# MUltiplico por 1e9 la seguda columna de ambos datos
g_reset_data[:, 1] *= 1e9
g_set_data[:, 1] *= 1e9
```

```python
def config_ax(ax):
    ax.grid(which="major", color="#DDDDDD", linewidth=0.8, zorder=-1)
    ax.grid(which="minor", color="#DEDEDE", linestyle=":", linewidth=0.5, zorder=-1)
    ax.minorticks_on()
    ax.tick_params(axis="both", which="both", direction="in", top=True, right=True)


def setup_plt(plt, latex=True, scaling=1):
    plt.rcParams.update(
        {
            "pgf.texsystem": "pdflatex",
            "text.usetex": latex,
            "font.family": "fourier",
            "text.latex.preamble": "\n".join(
                [
                    r"\usepackage[utf8]{inputenc}",
                    r"\usepackage[T1]{fontenc}",
                    r"\usepackage{siunitx}",
                ]
            ),
        }
    )

    SMALL_SIZE = 8 * scaling
    MEDIUM_SIZE = 10 * scaling
    BIGGER_SIZE = 11 * scaling

    plt.rc("font", size=SMALL_SIZE)
    plt.rc("axes", titlesize=SMALL_SIZE)
    plt.rc("axes", labelsize=MEDIUM_SIZE)
    plt.rc("xtick", labelsize=SMALL_SIZE)
    plt.rc("ytick", labelsize=SMALL_SIZE)
    plt.rc("legend", fontsize=SMALL_SIZE)
    plt.rc("figure", titlesize=BIGGER_SIZE)
    plt.rc("axes", titlesize=BIGGER_SIZE * 1.05)


# === Configuración de estilo LaTeX y plot ===
setup_plt(plt, latex=True, scaling=2)
```

```python
# Create figure with two subplots
fig1, ax1 = plt.subplots(1, 1, figsize=(12, 9))
fig2, ax2 = plt.subplots(1, 1, figsize=(12, 9))

config_ax(ax1)
config_ax(ax2)

figure_path_1 = r"C:/Users/jimdo/Documents/GitHub/RRAM_Simulation/ADS/g_reset_vs_V.png"

figure_path_2 = r"C:/Users/jimdo/Documents/GitHub/RRAM_Simulation/ADS/g_set_vs_V.png"

# Plot Reset data
ax1.plot(g_reset_data[:, 0], g_reset_data[:, 1], color="C1")
ax1.set(
    xlabel=r"Voltage (\SI{}{\volt})",
    ylabel=r"$g$ reset (\SI{}{\nano\meter})",
            # title=f"Sim {num}: g completa (set)",
)
fig1.savefig(figure_path_1, dpi=300, bbox_inches="tight")
plt.close(fig1)
        
# Plot Set data
ax2.plot(g_set_data[:, 0], g_set_data[:, 1], color="C3")
ax2.set(
    xlabel=r"Voltage (\SI{}{\volt})",
    ylabel=r"$g$ set (\SI{}{\nano\meter})",
    # title=f"Sim {num}: g completa (reset)",
)
fig2.savefig(figure_path_2, dpi=300, bbox_inches="tight")
plt.close(fig2)
```
