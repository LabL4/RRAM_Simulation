```python
import cv2
import subprocess
import sys

# ruta_raiz = 'C:/Users/Usuario/Documents/GitHub/RRAM_Simulation/'
ruta_raiz = '/Users/antonio_lopez_torres/Documents/GitHub/RRAM_Simulation/' # Ruta en el mac
sys.path.append(ruta_raiz)
```

```python
ruta_oxigenos_data = ruta_raiz + 'Results/reset/Oxygen_pp_reset_0.pkl'
ruta_configuracion_data = ruta_raiz + 'Results/reset/Configurations_pp_reset_0.pkl'

ruta_oxigenos_save = ruta_raiz + 'Videos/Oxygen_pp_reset_0.mp4'
ruta_configuracion_save = ruta_raiz + 'Videos/Configurations_pp_reset_0.mp4'

ruta_simulation_save = ruta_raiz + 'Videos/Simulacion_sistema_reset.mp4'

ruta_script_configuration = ruta_raiz + 'Videos/ReprePal_configuracion.py'
ruta_script_oxigen = ruta_raiz + 'Videos/ReprePal_oxigeno.py'
```

```python
subprocess.run(["python", ruta_script_configuration, ruta_configuracion_data, ruta_configuracion_save])
# subprocess.run(["python", ruta_script_oxigen, ruta_oxigenos_data, ruta_oxigenos_save])
```

```python
# Cargar los videos
video1 = cv2.VideoCapture(ruta_configuracion_save)
video2 = cv2.VideoCapture(ruta_oxigenos_save)

# Obtener el tamaño del video
frame_width = int(video1.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(video1.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(video1.get(cv2.CAP_PROP_FPS))

# Crear un objeto VideoWriter para guardar el video resultante
out = cv2.VideoWriter(ruta_simulation_save, cv2.VideoWriter_fourcc(*'mp4v'),
                      fps, (frame_width, frame_height))

while True:
    ret1, frame1 = video1.read()
    ret2, frame2 = video2.read()

    if not ret1 or not ret2:
        break

    # Normalizar los frames de ambos videos
    frame1 = cv2.normalize(frame1, None, 0, 255, cv2.NORM_MINMAX)
    frame2 = cv2.normalize(frame2, None, 0, 255, cv2.NORM_MINMAX)

    # Superponer los frames
    blended_frame = cv2.addWeighted(frame1, 0.5, frame2, 0.5, 0)

    # Escribir el frame resultante
    out.write(blended_frame)

# Liberar los objetos
video1.release()
video2.release()
out.release()

cv2.destroyAllWindows()
```
