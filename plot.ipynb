```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Rango de simulaciones
total_simulaciones = 9  # <-- cámbialo al total que tengas

for num_simulation in range(1, total_simulaciones + 1):
    simulation_path = os.path.join(ruta_raiz, 'Results', f'simulation_{num_simulation}/')
    # Rutas relativas a cada figura
    imagenes = [
        simulation_path + "Figures/" + f"initial_pp_set_{num_simulation-1}.png",
        simulation_path + "Figures/" + f"Final_state_pp_set_{num_simulation}.png",
        simulation_path + "Figures/" + f"Final_state_sp_set_{num_simulation}.png",
        simulation_path + "Figures/" + f"Final_state_pp_reset_{num_simulation}.png",
        simulation_path + "Figures/" + f"Final_state_pp_reset_{num_simulation}.png",
    ]

    # Crear figura con 2 filas y 3 columnas
    fig, axes = plt.subplots(2, 3, figsize=(15, 8))
    axes = axes.flatten()

    for i, ax in enumerate(axes):
        if i < len(imagenes):
            img_path = imagenes[i]
            if os.path.exists(img_path):  # 👈 Evita error si imagen no existe
                img = mpimg.imread(img_path)
                ax.imshow(img)
                nombre_base = os.path.splitext(os.path.basename(img_path))[0]
                ax.set_title(nombre_base, fontsize=12)
            else:
                ax.text(0.5, 0.5, "Imagen no encontrada", ha="center", va="center", fontsize=10)
            ax.axis('off')
        else:
            ax.axis('off')

    plt.tight_layout()
    output_path = figures_path + f"State_grupal_{num_simulation}.png"
    plt.savefig(output_path, dpi=300)
    plt.close()  # evita mostrar múltiples figuras en notebook

    print(f"✅ Figura grupal guardada: {output_path}")
```
