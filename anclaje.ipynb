```python
import numpy as np

# Creamos una matriz de ejemplo: 50 filas (alto) por 100 columnas (ancho)

matriz_ejemplo = np.zeros((50, 100))

# Extraemos los valores usando .shape
num_filas, num_columnas = matriz_ejemplo.shape

print(f"Filas (Eje Y): {num_filas}")
print(f"Columnas (Eje X): {num_columnas}")
```
