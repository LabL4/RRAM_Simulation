```python
import cv2
```

```python

# Cargar los videos
video1 = cv2.VideoCapture('Configuracion.mp4')
video2 = cv2.VideoCapture('Oxygen.mp4')

# Obtener el tamaño del video
frame_width = int(video1.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(video1.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(video1.get(cv2.CAP_PROP_FPS))

# Crear un objeto VideoWriter para guardar el video resultante
out = cv2.VideoWriter('output.mp4', cv2.VideoWriter_fourcc(*'mp4v'), fps, (frame_width, frame_height))

while True:
    ret1, frame1 = video1.read()
    ret2, frame2 = video2.read()

    if not ret1 or not ret2:
        break

    # Superponer los frames
    blended_frame = cv2.addWeighted(frame1, 0.5, frame2, 0.65, 0)

    # Escribir el frame resultante
    out.write(blended_frame)

# Liberar los objetos
video1.release()
video2.release()
out.release()
cv2.destroyAllWindows()
```
